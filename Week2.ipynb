```python
#get the items not common to both series A and series B
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
ser_un = pd.Series(np.union1d(ser1, ser2))
ser_in = pd.Series(np.intersect1d(ser1, ser2))
print(ser_un[~ser_un.isin(ser_in)])
```

# MATPLOTLIB

```python
#Show the relationship between Total profit and no. of month using Line Plot
plt.plot(data['month_number'],data['total_profit'], '-ob', mfc='hotpink', ms=10)
plt.xlabel("Month number")
plt.ylabel("Profit")
plt.legend(['Profit Line'])
```

```python
#Read all product sales data and show it using a multi line plot. Display the number of units sold per month for each product using multi line plots
plt.plot(data['month_number'], data['bathingsoap'],'-o', mfc='hotpink', lw=3, label="Bathingsoap Sales data")
plt.plot(data['month_number'], data['facecream'],'-o', mfc='c', lw=3, label="Facecream Sales data")
plt.plot(data['month_number'], data['facewash'], '-o', mfc='y', lw=3, label="Facecream Sales data")
plt.plot(data['month_number'], data['shampoo'], '-o', mfc='c', lw=3, label="Shampoo Sales data")
plt.plot(data['month_number'], data['toothpaste'],'-o', mfc='#4CAF50', lw=3, label="Toothpaste Sales data")
plt.xlabel("Month number")
plt.ylabel("Sales unit in number")
plt.title("SALES DATA")
plt.legend(loc="upper left")
```

```python
#Plot a scatter plot between no. of month and toothpaste units sold
plt.scatter(data['month_number'], data['toothpaste'], label="Toothpaste sales")
plt.xlabel("Month number")
plt.ylabel("toothpaste units sold")
plt.title("Toothpaste Sales Data")
plt.legend(loc="upper left")
```

```python
#Add a gridline to the scatter plot
plt.scatter(data['month_number'], data['toothpaste'], label="Toothpaste sales")
plt.xlabel("Month number")
plt.ylabel("toothpaste units sold")
plt.title("Toothpaste Sales Data")
plt.legend(loc="upper left")
plt.grid(color = 'g', linestyle = '--', linewidth = 0.5)
```

```python
#Plot a bar chart which displays the number of units sold per month for face cream and face wash. Add a separate bar for each product in the same chart.(with grid lines)
plt.bar([a-0.25 for a in data['month_number']], data['facecream'], label="Facecream sales", width=0.35)
plt.bar([a+0.5 for a in data['month_number']],data['facewash'], label="Facewash sales", width=0.35)
plt.xlabel("Month number")
plt.ylabel("units sold")
plt.title("Facecream & Facewash Sales Data")
plt.legend(loc="upper left")
plt.grid(color = 'g', linestyle = '--', linewidth = 0.5)
```

```python
#Plot the total profit of each month and show it using the histogram to see the most common profit ranges
range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist (list(data["total_profit"]), range, label='Profit data')
plt.xlabel('profit Range')
plt.ylabel('Actual Profit')
plt.legend (loc='upper left')
plt.xticks (range)
plt.title("Profit data")
plt.show()
```

```python
#Plot a pie chart for total sales data
labels = ['FaceCream', 'FaceWash', "ToothPaste", "Bathing soap", "Shampoo", "Moisturizer"]
salesData = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(), data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend (loc='lower left')
plt.title('Sales Data')
plt.show()
```

```python
#Read all product sales data and show it using the stack plot.
plt.plot([],[],color='r', label='face Cream', linewidth=5)
plt.plot([],[],color='k', label='Face wash', linewidth=5)
plt.plot([],[],color='hotpink', label='Tooth paste', linewidth=5)
plt.plot([],[],color='c', label='Bathing soap', linewidth=5)
plt.plot([],[],color='g', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)
plt.stackplot (data['month_number'], data['facecream'], data['facewash'], data['toothpaste'], data['bathingsoap'], data['shampoo'], data['moisturizer'],
               colors=['r', 'k', 'hotpink', 'c', 'g', 'y',])                                                                                                                                                              
plt.xlabel("Month Number")
plt.ylabel('Sales unints in Number')
plt.title('All product sales data using stack plot')
plt.legend (loc='upper left')
plt.show()
```
